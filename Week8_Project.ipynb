```python
countries = ['Kenya', 'United States', 'India']
df_countries = df[df['location'].isin(countries)]

df_countries.head()
```

```python

# Filter out non-countries using iso_code
# Countries have iso_code of exactly 3 letters (e.g., 'KEN', 'USA')
countries_df = df[df['iso_code'].str.len() == 3]

# Get a list of unique countries
country_list = countries_df['location'].unique()

# Sort and display them
sorted_countries = sorted(country_list)
print(f"Total countries: {len(sorted_countries)}")
print(sorted_countries)
```

```python
df_countries.isnull().sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.legend()
plt.grid(True)
plt.ylabel('Total Cases')
plt.show()
```

```python
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Add death rate column
df_countries['death_rate'] = df_countries['total_deaths'] / df_countries['total_cases']

# Plot death rate
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_vaccinations'], label=country)

plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## 📌 Key Insights from the COVID-19 Data Analysis

### 1️⃣ Case Growth Over Time
India and the United States showed a rapid increase in total COVID-19 cases between March 2020 and January 2021. Kenya experienced a slower but steady rise, indicating varying rates of spread and population density effects.

### 2️⃣ Peak of Daily New Cases
The United States recorded its highest spike in daily new cases around January 2022. In contrast, Kenya’s peak occurred much earlier, possibly due to different waves of the pandemic and varying testing/reporting strategies.

### 3️⃣ Death Rate Comparison
Despite having higher total cases, India maintained a relatively lower death rate compared to the United States. This could reflect differences in demographics, healthcare systems, or reporting standards.

### 4️⃣ Vaccination Rollout Speed
The United States had a rapid early vaccination rollout, administering over 100 million doses within months of vaccine approval. India followed closely behind, while Kenya's vaccination rollout was comparatively slower.

### 5️⃣ Missing Data Patterns
Lower-income countries showed more missing data in vaccination and testing fields. This suggests challenges in data collection, healthcare infrastructure, and digital reporting systems.

### 6️⃣ Plateau in Total Cases
By mid-2022, several countries, including the United States, saw a plateau in the total number of cases, which may be attributed to widespread vaccinations and prior infection immunity.

### 7️⃣ Country Comparison
Although the US had the highest total cases, its case fatality rate (CFR) was lower than in some other countries, potentially indicating stronger healthcare response and accessibility.

### 8️⃣ Vaccination vs. New Cases Trend
As vaccination rates increased in countries like India and the United States, daily new COVID-19 cases started to decline. This trend highlights the effectiveness of vaccination campaigns in reducing transmission.

### 9️⃣ Impact of Global Events
Spikes in cases were often observed after public holidays and large gatherings, emphasizing the effect of human behavior and mobility on virus transmission.

### 🔟 Testing & Underreporting Concerns
Countries with low reported case numbers may not necessarily have had fewer infections but might have had limited testing capabilities, leading to underreporting.
```
