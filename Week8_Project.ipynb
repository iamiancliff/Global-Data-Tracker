```python
countries = ['Kenya', 'United States', 'India']
df_countries = df[df['location'].isin(countries)]

df_countries.head()
```

```python

# Filter out non-countries using iso_code
# Countries have iso_code of exactly 3 letters (e.g., 'KEN', 'USA')
countries_df = df[df['iso_code'].str.len() == 3]

# Get a list of unique countries
country_list = countries_df['location'].unique()

# Sort and display them
sorted_countries = sorted(country_list)
print(f"Total countries: {len(sorted_countries)}")
print(sorted_countries)
```

```python
df_countries.isnull().sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.legend()
plt.grid(True)
plt.ylabel('Total Cases')
plt.show()
```

```python
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Add death rate column
df_countries['death_rate'] = df_countries['total_deaths'] / df_countries['total_cases']

# Plot death rate
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))

for country in countries:
    subset = df_countries[df_countries['location'] == country]
    plt.plot(subset['date'], subset['total_vaccinations'], label=country)

plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.grid(True)
plt.show()
```
